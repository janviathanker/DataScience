```python
# Program 1(a): Tossing a Coin 10,000 Times
import random

trials = 10000
heads = 0
tails = 0

for _ in range(trials):
    toss = random.choice(['H', 'T'])
    if toss == 'H':
        heads += 1
    else:
        tails += 1

p_heads = heads / trials
p_tails = tails / trials

print("Coin Toss Simulation (10,000 tosses)")
print(f"Probability of Heads ≈ {p_heads:.4f}")
print(f"Probability of Tails ≈ {p_tails:.4f}")
```

```python
# Program 1(b): Rolling Two Dice and Probability of Getting a Sum of 7
import random

trials = 10000
sum_7 = 0

for _ in range(trials):
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    if die1 + die2 == 7:
        sum_7 += 1

p_sum_7 = sum_7 / trials

print("Rolling Two Dice (10,000 trials)")
print(f"Probability of getting a sum of 7 ≈ {p_sum_7:.4f}")
```

```python
# Program 2: Probability of Getting at Least One "6" in 10 Rolls
import random

def prob_at_least_one_six(trials=10000, rolls=10):
    success = 0
    for _ in range(trials):
        has_six = False
        for _ in range(rolls):
            if random.randint(1, 6) == 6:
                has_six = True
                break
        if has_six:
            success += 1
    return success / trials

p_one_six = prob_at_least_one_six()
print("Probability of getting at least one '6' in 10 rolls:")
print(f"Estimated Probability ≈ {p_one_six:.4f}")
```

```python
# Program 3: Conditional Probability and Bayes’ Theorem
import numpy as np

colors = ['red', 'green', 'blue']
probabilities = [5/20, 7/20, 8/20]

n = 1000
draws = np.random.choice(colors, size=n, p=probabilities)

# Calculate P(Red | Previous Blue)
count_prev_blue = 0
count_red_given_blue = 0

for i in range(1, n):
    if draws[i-1] == 'blue':
        count_prev_blue += 1
        if draws[i] == 'red':
            count_red_given_blue += 1

p_red_given_blue = count_red_given_blue / count_prev_blue

# Individual probabilities
p_red = probabilities[0]
p_blue = probabilities[2]

# Estimate P(Blue | Red)
count_red = np.sum(draws == 'red')
count_blue_given_red = 0
for i in range(1, n):
    if draws[i-1] == 'red' and draws[i] == 'blue':
        count_blue_given_red += 1

p_blue_given_red = count_blue_given_red / count_red

# Bayes’ Theorem
p_red_from_bayes = (p_blue_given_red * p_red) / p_blue

print("Conditional Probability and Bayes’ Theorem")
print(f"Estimated P(Red | Previous Blue) ≈ {p_red_given_blue:.4f}")
print(f"Bayes’ Theorem P(Red | Blue) ≈ {p_red_from_bayes:.4f}")
```

```python
# Program 4: Discrete Random Variable Statistics
import numpy as np

values = [1, 2, 3]
probs = [0.25, 0.35, 0.4]

sample = np.random.choice(values, size=1000, p=probs)

mean_emp = np.mean(sample)
var_emp = np.var(sample)
std_emp = np.std(sample)

print("Discrete Random Variable Statistics")
print(f"Empirical Mean ≈ {mean_emp:.4f}")
print(f"Empirical Variance ≈ {var_emp:.4f}")
print(f"Empirical Std. Dev ≈ {std_emp:.4f}")
```

```python
# Program 5: Continuous Random Variable (Exponential Distribution)
import numpy as np
import matplotlib.pyplot as plt

data = np.random.exponential(scale=5, size=2000)

plt.figure(figsize=(8,5))
plt.hist(data, bins=30, density=True, alpha=0.6, label='Histogram')

# Overlay PDF
x = np.linspace(0, max(data), 100)
pdf = (1/5) * np.exp(-x/5)
plt.plot(x, pdf, 'r-', label='PDF Overlay')

plt.title("Exponential Distribution (mean = 5)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
# Program 6: Central Limit Theorem Simulation
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
population = np.random.uniform(0, 1, 10000)

sample_means = []
for _ in range(1000):
    sample = np.random.choice(population, size=30)
    sample_means.append(np.mean(sample))

# Plot the original uniform distribution
plt.figure(figsize=(8,5))
plt.hist(population, bins=30, alpha=0.6, density=True, label='Uniform Distribution')
plt.title("Uniform Distribution (Population)")
plt.legend()
plt.show()

# Plot the distribution of sample means
plt.figure(figsize=(8,5))
plt.hist(sample_means, bins=30, alpha=0.6, density=True, label='Sample Means')
plt.title("Central Limit Theorem: Distribution of Sample Means")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.legend()
plt.show()
```
